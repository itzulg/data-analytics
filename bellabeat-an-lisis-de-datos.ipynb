## **Análisis No.1: Pasos diarios VS Calorías diarias.**
> * Es necesario analizar los pasos diarios de cada usuario, para saber más acerca de los hábitos de ejercicio que realizan.
> * Se saca el **promedio de pasos diarios y calorías diarias**, esto con el motivo de ver si los usuarios estan quemando esas calorías con una cantidad de actividad física adecuada. 
> * Para sacar el promedio de los pasos diarios en cada usuario así como de sus calorías promedio, se usa la función **mean()**, esta función acepta un vector como entrada y devuelve el promedio como numérico.
> * La función **summarize()** reduce un marco de datos a un resumen de un solo vector o valor. 
> * Estos promedios se agrupan por medios de el **id**.

```r
#Promedio de pasos y calorías diarias
promediopyc <- pasoscalorias %>%
group_by (id) %>% #Se agrupan por medio del id

#Los promedios se resumen en una tabla
    summarize (prompasos = mean(steptotal), promcalorias = mean(calories))
head(promediopyc)
```

> **Visualización**.
>
> Se procede a realizar dos gráficos de dispersión:
> 1.  Gráfico de dispersión de los **pasos y calorías diarias.**
> 2. Gráfico de dispersión de los **pasos promedio y calorías promedio diarias.**

```r
#Gráfico de dispersión de los pasos y calorías diarias.
ggplot(data=promediopyc, aes(x=prompasos, y=promcalorias)) + 
  geom_point()  +  geom_smooth() + 
    labs(title="Promedio pasos diarios VS Promedio calorias diarias", 
                    subtitle='Pasos promedio totales en contraste con las calorías promedio diarias registradas al día',
                    x='Pasos promedio totales', y='Calorías promedio totales')+
 theme(plot.title=element_text(size=25, face='bold'),
      plot.subtitle=element_text(size=15))

#Gráfico de dispersión de los pasos promedio y calorías promedio diarias.
ggplot(data=pasoscalorias, aes(x=steptotal, y=calories)) + 
  geom_point()  +  geom_smooth() +
    labs(title="Total pasos VS Total Calorías", 
                    subtitle='Pasos totales en contraste con las calorías diarias registradas al día',
                    x='Pasos totales', y='Calorías totales')+
 theme(plot.title=element_text(size=25, face='bold'),
      plot.subtitle=element_text(size=15))
```

> **Observaciones**
>* Existe una **correlación positiva**, entre más pasos se realizan, más calorías se queman.

## **Análisis No.2: Clasificación de la actividad física de acuerdo a los pasos promedio diarios.**
Es necesario confirmar información del análisis anterior. De acuerdo al siguiente artículo científico de **Crespo-Salgado JJ, et al. "Guía básica de detección del sedentarismo y recomendaciones de actividad física en atención primaria". Aten Primaria. (2014).** [http://dx.doi.org/10.1016/j.aprim.2014.09.004](http://dx.doi.org/10.1016/j.aprim.2014.09.004), la actividad física puede ser clasificada de acuerdo a las siguienets categorías:
> *     **Sedentario o inactivo** - Menos de 5000 pasos al día.
> *     **Algo activo** - Realiza entre 5.000 y 9.999 pasos / día.
> *     **Activo** - Realiza entre 10.000 y 12.499 pasos / día. Cumple las recomendaciones de pasos / día.
> *     **Muy activo** - Realiza 12.500 pasos / día o más Cumple las recomendaciones de pasos / día.

Se usa esta clasificación en nuestro conjunto de datos por medio de el **promedio de pasos totales al día** que realizan los usuarios. Los resultados de la clasificación se pueden visualizar en un **gráfico circular**.

```r
#Clasificación de la actividad física
tipoact <- promediopyc %>%
  mutate( tipo_act= case_when( 
    prompasos < 5000 ~ "4.Sedentario",
    prompasos >= 5000 & prompasos < 9999 ~ "3.Algo activo", 
    prompasos >= 10000 & prompasos < 12599 ~ "2.Activo", 
    prompasos >= 12500 ~ "1.Muy activo"
  ))
 head(tipoact)
```

> El resultado de la clasificación de actividad física se convierte en porcentaje y este se almacena en una nueva columna por medio del uso de la función **mutate()** y visualizamos los cálculos de cada actividad física.

```r
#Tranformamos a procentajes
porc_tipoact <- tipoact %>%
  group_by(tipo_act)%>%
     summarize(total = n())%>% #Total igual al número de filas de cada categoría
        mutate(porc = round((total / sum(total)),4))%>%
        mutate( porcentaje= round(porc*100,2)) 
head(porc_tipoact) 
```

> **Visuaización.**
> 
> Se procede a realizar un gráfico circular para visualizar los porcentajes de cada categoría de actividad física, es importante realizar una visualización que sea **accesible para todo tipo de público**.

```r
#Gráfico circular
ggplot(porc_tipoact, aes(x="", y= porcentaje, fill=tipo_act)) +
    geom_bar(stat = "identity")+
    coord_polar( "y", start=0) + theme_minimal()+
   geom_text(aes(label = porcentaje),position = position_stack(vjust = 0.4)) +
    labs(title = "Niveles de actividad física", fill = "Nivel de actividad", 
         subtitle='Categorización de la actividad física de la muestra de FitBit Fitness Tracker',
         caption='Artículo: 2015;47(3):175-183')+
    theme(plot.title=element_text(size=25, face='bold'),
          plot.subtitle=element_text(size=15),
          plot.caption=element_text(size=12),
          axis.title.x = element_blank(),
          axis.title.y = element_blank(),
          panel.grid = element_blank()) +
scale_fill_manual(values = c("#23BAC4","#EF280F", "#E7D40A", "#E6EBE0"))
```

**Observaciones**
> * La mayoría de los participantes se encuentra en el nivel **"Algo activo"**, seguido de **"Sedentario", "Activo" y "Muy activo"**.
> * Se ve reflejado una **falta de actividad física en el día** en la mayoría de usuarios.
> * Se debe aumentar los niveles de actividad física al día para prevenir **enfermedades crónicas y para mejorar la calidad de vida.**

## **Análisis No.3: Tiempo de dormir VS Actividad en el día.**
> * Una de las cuestiones de interés en este análisis de datos es conocer los **hábitos** de los participantes,debido a esto, es necesario determinar si la poca actividad física es resultado de los patrones de sueño.

```r
#Promedio de los minutos de sueño y el promedio de los minutos de sedentarismo
promediodya <- dormiractividad %>%
group_by (id) %>%
    summarize (promdormir = mean(totalminutesasleep), promsed = mean(sedentaryminutes))

head(promediodya)
```

> **Visualización**
> 
> Se procede a crear un gráfico de dispersión para determinar si hay alguna relación entre los datos.

```r
#Gráfico de dispersión
ggplot(data=promediodya, aes(x=promdormir, y=promsed)) + 
   geom_point() + geom_smooth() + 
labs(title="¿El sueño tiene que ver con \n el sedentarismo?", 
     subtitle='Promedio de los minutos de sueño en contraste con el promedio de \n los minutos de actividad sedentaria al día',
     x='Promedio minutos de sueño', y='Promedio minutos sedentarios')+
 theme(plot.title=element_text(size=25, face='bold'), 
       plot.subtitle=element_text(size=15))
```

## **Análisis No.4: METs**
> * El **valor MET** de una actividad es calculado respecto al reposo. Los valores MET muestran la intensidades de una actividad física.
> * En este análisis es importarte centrarse en la intensidad de actividades que están realizando los usuarios para saber sus **preferencias de ejercicio.**
> * En la página web [https://www.hospitallamoraleja.es/es/pacientes-mujer-deporte-capitulo14.php](https://www.hospitallamoraleja.es/es/pacientes-mujer-deporte-capitulo14.php) se proporciona una tabla donde expone la **clasificación en METs de algunas actividades comunes.**




> * Dentro del conjunto de datos encontramos un error, **el punto decimal se encuentra recorrido un espacio hacia la derecha**, esta corrección está respaldada en este [foro](http://https://community.fitbit.com/t5/Web-API-Development/Definition-of-Mets-in-Charge-HR/td-p/1240824) de la comunidad de FitBit.
>
> * Para corregir, es necesario dividir entre 10 la columa de mets.

```r
met<-met%>%
mutate(met,mets=(mets/10))
head(met)
```

> * Se realizará un análisis de los **METs promedio** en los usuarios durante los días de la semana.

```r
promediomets <- met %>%
 separate(date, into = c("date", "time"), sep= " ") %>% #separamos la fecha y la hora
  mutate(date = ymd(date)) %>% #formato de fecha
  mutate(semana = weekdays(date)) #nueva columna que especifica el día de la semana

promediomets$semana <-ordered(promediomets$semana, levels=c(
    "Monday", "Tuesday", "Wednesday", "Thursday",
    "Friday", "Saturday", "Sunday")) #Ordenamos los días de la semana

promediomets <- promediomets%>%
  group_by(semana)%>%
    summarize (promet = round(mean(mets),3)) #promedio por día de los METs
        

View(promediomets)
```

>**Visualización**
>
>Crearemos un gráfico de barras para comparar los **METs promedio** por día de la semana.


```r
ggplot(promediomets) +
      geom_col(aes(semana, promet), fill = "#23BAC4") +
      labs(title = "METs a la semana", x= "", y = "METs promedio", 
           subtitle="METs promedio de los usuarios en una semana") +
      theme(plot.title=element_text(size=25, face='bold'),
            plot.subtitle=element_text(size=15),
            axis.text.x = element_text(angle = 45,vjust = 1, hjust = 1,size=12),
            axis.text.y = element_text(size=15))
```

> **Observaviones**
>
>* Todos los usuarios realizan **actividades en un nivel leve**.
>* Este análisis confirma el análisis realizado con anterioridad sobre la categorización de actividad física de acuerdo a sus pasos.
